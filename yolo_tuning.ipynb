```python
from ultralytics import YOLO
import torch

model = YOLO("yolov8n.pt")

model.tune(
    data="yolo.yaml",
    epochs=30,        # shorter for tuning
    iterations=30,    # good balance for your dataset size
    imgsz=640,
    device=device,
    plots=True,
    save=True
)
```

Training

```python
model.train(
    data="yolo.yaml",
    epochs=100,
    imgsz=640,
    batch=16 if device != "cpu" else 8,
    device=device,
    patience=10,   # amount of epochs before stopping
    workers=4
)
```

Visualizations

```python
from ultralytics import YOLO

model = YOLO("best.pt")

metrics = model.val() 

metrics.curves
metrics.curves_results
metrics.confusion_matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np

recall = np.array(metrics.curves_results[0][0]).squeeze()
precision = np.array(metrics.curves_results[0][1]).squeeze()
print(recall, precision)
plt.figure()
plt.plot(recall, precision, linewidth=2, color='blue')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curve")
plt.savefig('pr_curve.jpg')
plt.grid()


plt.show()
```
